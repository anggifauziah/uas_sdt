## Correlation

```python
cols = dataset.corr()
print(cols['price_range'].sort_values(ascending = False),'\n')
```

```python
k = 21 #Jumlah variabel
cols = dataset.corr().nlargest(k, 'price_range')['price_range'].index
cm = dataset[cols].corr()
plt.figure(figsize=(30,20))
sns.heatmap(cm, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='viridis',
            linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
plt.show()
```

# Data Visualization

## Data distribution

```python
feature = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']

plt.figure(figsize=(20,10))
count = 0

for i in feature:
    myexplode = [0.2, 0]
    
    colors = ['#66b3ff','#ff9999','#ff9999','#ffcc99']
    
    labels = dataset[i].value_counts().index
    sizes = dataset[i].value_counts().values  
    
    plt.subplot(2,3,count+1)
    plt.pie(sizes, labels=labels, explode = myexplode, shadow = True, startangle=90, colors=colors, autopct='%1.1f%%')
    plt.title(f'Distribution of {feature[count]}', color = 'black',fontsize = 15)
    count += 1
```

## No of Phones vs Camera megapixels of front and primary camera

```python
plt.figure(figsize=(10,6))
dataset['fc'].hist(alpha=0.5,color='blue',label='Front camera')
dataset['pc'].hist(alpha=0.5,color='red',label='Primary camera')
plt.legend()
plt.xlabel('MegaPixels')
```
